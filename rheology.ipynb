# Pow law rheology

```python
import rheology
import numpy as np
import pandas

rock = pandas.read_csv('rock.csv')
rock
```

# Hirth and Kohlstedt, 2003, dry olivine, dislocation creep
## HK03dd

```python
i      = 0
T      = np.arange(600,1200)+273
strain = 1e-12
rhe    = rheology.flow_law(rock.A[i], rock.n[i], rock.d[i], rock.m[i], rock.Coh[i], 
                           rock.r[i], rock.Q[i], 1e9, rock.V[i], T, epsilon=strain)
rhe.plot_T_eta(T, rhe.viscosity()*1e6)
#rhe.plot_T_stress(T, rhe.GetSigma())
```
