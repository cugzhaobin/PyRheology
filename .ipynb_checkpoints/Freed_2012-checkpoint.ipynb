## Freed, 2012. Using short-term postseismic displacements to infer the ambient deformation conditoions of the upper mantle

```python
import pandas as pd
import calc_temperature
import numpy as np
import rheology
import matplotlib.pyplot as plt

def plot_rheology(vis1, vis2, dep, T, label_1="", label_2=""):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 6), dpi=300)
    ax1.plot(vis1, dep, label=label_1)
    ax1.plot(vis2, dep, label=label_2)
    ax1.set_xscale('log')
    ax1.set_ylim([30,100])
    ax1.set_xlim([1e17,1e24])
    ax1.legend()
    ax1.set_xlabel('Viscosity (Pa.s)')
    ax1.set_ylabel('Depth (km)')
    ax1.grid(True, which="both", ls="--", alpha=0.5)
    ax1.invert_yaxis()
    
    ax2.plot(T, dep, color="red")
    ax2.set_xlim(min(T), max(T))
    ax2.set_xlabel("Temperature (K)")
    ax2.grid(True, ls="--", alpha=0.5)
    ax2.set_ylim([30,100])
    ax2.invert_yaxis()
    plt.tight_layout()
    plt.show()
```

## Dislocation: Sensitive to background strain rate

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
print(f'{max(P):e}')
T     = calc_temperature.calc_T(depth, 273, 1200+273, 15, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30, 3.5, 0, 0, 1000, 1.2, 480e3, P[idx], 11e-6, T[idx], epsilon=3.175e-12)
rh2   = rheology.flow_law(30, 3.5, 0, 0, 1000, 1.2, 480e3, P[idx], 11e-6, T[idx], epsilon=3.175e-13)
plot_rheology(rh1.get_viscosity(), rh2.get_viscosity(), depth[idx], T[idx], label_1="1e-7/year", label_2="1e-8/year")
```

## Dislocation: Sensitive to stress exponent

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 14, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30, 3.2, 0, 0, 1000, 1.2, 480e3, P[idx], 11e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30, 3.8, 0, 0, 1000, 1.2, 480e3, P[idx], 11e-6, T[idx], epsilon=3e-15)
plot_rheology(rh1.get_viscosity(), rh2.get_viscosity(), depth[idx], T[idx], label_1="n=3.2", label_2="n=3.8")
```

## Sensitive to water content

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 14, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30e6, 3.5, 0, 0, 2000, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30e6, 3.5, 0, 0, 200, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
plot_rheology(rh1.get_viscosity()*1e6, rh2.get_viscosity()*1e6, depth[idx], T[idx], label_1="Coh=2000", label_2="Coh=1000")
```

## Sensitive to activation energy

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 14, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30e6, 3.5, 0, 0, 1000, 1.2, 440e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30e6, 3.5, 0, 0, 1000, 1.2, 520e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
plot_rheology(rh1.get_viscosity()*1e6, rh2.get_viscosity()*1e6, depth[idx], T[idx], label_1="Q=440", label_2="Q=520")
```

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 14, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30e6, 3.5, 0, 0, 1000, 1.2, 480e3, P[idx], 6e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30e6, 3.5, 0, 0, 1000, 1.2, 480e3, P[idx], 16e-6, T[idx], epsilon=3e-15)
plot_rheology(rh1.get_viscosity()*1e6, rh2.get_viscosity()*1e6, depth[idx], T[idx], label_1="V=6e-6", label_2="V=16e-6")
```
