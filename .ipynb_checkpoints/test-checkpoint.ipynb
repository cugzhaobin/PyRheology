```python
import numpy as np
import h5py, yaml
import sys, os
import calc_temperature
import byerlee

layers =[]


with open('structure_cfg.yaml', 'r') as fid:
    lines = fid.read()
    cfg   = yaml.load(lines, Loader=yaml.FullLoader)
    gen_dict = cfg['general_dict']
    for dic in cfg.keys():
        if dic.find('layer') == 0:
            layers.append(cfg[dic])

z = np.linspace(0,200,200)
Ts = 0+273
Ti = 1200+273
G  = 7
thid = 2
T1 = calc_temperature.calc_T(z, Ts, Ti, G, thid)
G=15
T2 = calc_temperature.calc_T(z, Ts, Ti, G, thid)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,6))
plt.plot(T1,z, label='Pamir')
plt.plot(T2,z, label='Tibet')
ax=plt.gca()
ax.invert_yaxis()
plt.xlabel('Temperature')
plt.ylabel('Depth(km)')
plt.legend()
```

```python
ztop = 0
zbot = 0
Pbot = 0
irh = 'PT92'
g   = 9.81
h5 = h5py.File('rock.h5', 'r')

for i in [0,1]:
    print('--------------------------')
    layer = layers[i]
    zbot = ztop + layer['thickness']
    Ptop = Pbot
    irock = layer['rock_id']
    irh  = layer['rheo']
    density = h5['{}/{}/rkdensity'.format(irock, irh)][()]
    rhog = density*1e-3*g
    Pbot = Ptop + layer['thickness']*rhog
    
    did = 3
    Wk  = 1
    S,f=byerlee.byerlee(3,2)
    Pressure = Ptop+(z-ztop)*rhog;
    s    = S+f*Pressure*Wk
    print(irock, irh, density, rhog, Ptop, Pbot)
    

h5.close()   

plt.figure(figsize=(4,6))
plt.plot(s,z)
ax=plt.gca()
ax.invert_yaxis()
plt.xlabel('Stress')
plt.ylabel('Depth(km)')
```

## Freed, 2012. Using short-term postseismic displacements to infer the ambient deformation conditoions of the upper mantle

```python
import pandas as pd
import calc_temperature
import numpy as np
import rheology
import matplotlib.pyplot as plt

def plot_rheology(vis1, vis2, dep, T, label_1="", label_2=""):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    ax = plt.gca()
    plt.plot(vis1, dep, label=label_1)
    plt.plot(vis2, dep, label=label_2)
    ax.set_xscale('log')
    
    
    plt.ylim([30,100])
    plt.xlim([1e17,1e24])
    plt.legend()
    ax.invert_yaxis()
    
    plt.subplot(1,2,2)
    ax = plt.gca()
    ax.plot(T, dep)
    ax.invert_yaxis()
    ax.set_xlabel('Viscosity (Pa.s)')
    ax.set_ylabel('Depth (km)')
    ax.set_xscale('log')
    
    plt.subplot(1,2,2)
    ax = plt.gca()
    ax.plot(T, dep)
    ax.set_xlabel('Temperature (K)')
    ax.set_ylabel('Depth (km)') 
    ax.set_xscale('linear')
    #x.invert_yaxis()
    plt.show()
```

## Sensitive to background strain rate

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 13, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30e6, 3.5, 0, 0, 1000, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30e6, 3.5, 0, 0, 1000, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-16)
plot_rheology(rh1.viscosity()*1e6, rh2.viscosity()*1e6, depth[idx], T[idx], label_1="1e-7/year", label_2="1e-8/year")
```

## Sensitive to stress exponent

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 14, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30e6, 3.2, 0, 0, 1000, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30e6, 3.8, 0, 0, 1000, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
plot_rheology(rh1.viscosity()*1e6, rh2.viscosity()*1e6, depth[idx], T[idx], label_1="n=3.2", label_2="n=3.8")
```

## Sensitive to water content

```python
depth = np.arange(0,100,1)
P     = 3000 * 10 * depth * 1e3
T     = calc_temperature.calc_T(depth, 273, 1200+273, 14, 2)
idx   = np.where(depth>=30)[0]

rh1   = rheology.flow_law(30e6, 3.5, 0, 0, 2000, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
rh2   = rheology.flow_law(30e6, 3.5, 0, 0, 200, 1.2, 480e3, P[idx], 10e-6, T[idx], epsilon=3e-15)
plot_rheology(rh1.viscosity()*1e6, rh2.viscosity()*1e6, depth[idx], T[idx], label_1="Coh=2000", label_2="Coh=1000")
```
